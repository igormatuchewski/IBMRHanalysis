```python
import pandas as pd
import plotly.express as plt
```

```python
df9 = pd.read_csv('dataset/IBM_RH_dataset.csv')
df9.head(2)
```

```python
df9.rename(columns={
    'Attrition' : 'Unsatisfied'
}, inplace=True)

unsatisfied_mapping = {
    'Yes' : 'Unsatisfied',
    'No' : 'Satisfied'
}

df9['Unsatisfied'] = df9['Unsatisfied'].map(unsatisfied_mapping)

education_mapping = {
    1: 'Below College',
    2: 'College',
    3: 'Bachelor',
    4: 'Master',
    5: 'Doctor'
}

df9['Education'] = df9['Education'].map(education_mapping)

environment_mapping = {
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}

df9['EnvironmentSatisfaction'] = df9['EnvironmentSatisfaction'].map(environment_mapping)

jobinv_mapping = {
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}

df9['JobInvolvement'] = df9['JobInvolvement'].map(jobinv_mapping)

jobsatif_mapping = {
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}

df9['JobSatisfaction'] = df9['JobSatisfaction'].map(jobinv_mapping)

perfrating_mapping = {
    1: 'Low',
    2: 'Good',
    3: 'Excellent',
    4: 'Outstanding'
}

df9['PerformanceRating'] = df9['PerformanceRating'].map(perfrating_mapping)

relations_mapping = {
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}

df9['RelationshipSatisfaction'] = df9['RelationshipSatisfaction'].map(relations_mapping)

worklife_mapping = {
    1: 'Bad',
    2: 'Good',
    3: 'Better',
    4: 'Best'
}

df9['WorkLifeBalance'] = df9['WorkLifeBalance'].map(relations_mapping)

# Range distance from home
def distanceFromHome(distance):
    if 1 <= distance <= 5:
        return 'Close'
    elif 6 <= distance <= 12:
        return 'Medium'
    elif 13 <= distance <= 20:
        return 'Far Away'
    else:
        return 'Very Far Away'

df9['DistanceRange'] = df9['DistanceFromHome'].apply(distanceFromHome)

# Wage Range
def wageRange(wage):
    if 1000 <= wage < 5000:
        return '1K~5K'
    elif 5001 <= wage < 10000:
        return '5K~10K'
    elif 10001 <= wage < 15000:
        return '10K~15K'
    else:
        return '+15K'

df9['IncomeRange'] = df9['MonthlyIncome'].apply(wageRange)

# Age Range
def rangeAge(age):
    if 18 <= age <= 25:
        return '18~25'
    elif 26 <= age <= 35:
        return '26~35'
    elif 36 <= age <= 50:
        return '36~50'
    else:
        return '+50'
    
df9['AgeRange'] = df9['Age'].apply(rangeAge)

# YearsAtCompany Range
def yAtCRange(yAtC):
    if 0 <= yAtC <= 3:
        return '0~3'
    elif 4 <= yAtC <= 10:
        return '4~10'
    elif 11 <= yAtC <= 16:
        return '11~16'
    elif 17 <= yAtC <= 25:
        return '17~25'
    elif 26 <= yAtC <= 35:
        return '26~35'
    else:
        return '+36'

df9['YearsAtCompanyRange'] = df9['YearsAtCompany'].apply(yAtCRange)

# YearsInCurrentRole Range

def yInRRange(yInR):
    if 0 <= yInR <= 5:
        return '0~5'
    elif 6 <= yInR <= 10:
        return '6~10'
    elif 11 <= yInR <= 15:
        return '11~15'
    else:
        return '16+'
    
df9['YearsInCurrRoleRange'] = df9['YearsInCurrentRole'].apply(yInRRange)

# YearsWithCurrManager Range

def yWtMRange(yWtM):
    if 0 <= yWtM <= 5:
        return '0~5'
    elif 6 <= yWtM <= 10:
        return '6~10'
    elif 11 <= yWtM <= 15:
        return '11~15'
    else:
        return '16+'
    
df9['YearsWithCurrManagerRange'] = df9['YearsWithCurrManager'].apply(yWtMRange)

# YearsSinceLastPromotion Range

def ySlPRange(ySlP):
    if 0 <= ySlP <= 3:
        return '0~3'
    elif 4 <= ySlP <= 8:
        return '4~8'
    elif 9 <= ySlP <= 12:
        return '9~12'
    else:
        return '13+'
    
df9['YearsSinceLastPromotionRange'] = df9['YearsSinceLastPromotion'].apply(ySlPRange)

df9.head()
```

```python
df9.to_csv('df_ibm.csv', index=False)
```
